#7. Manipulaciones con arreglos.

Atributos: determinar size, shape, memory consumption data type of the array.

indexing arrays: getting and setting the value of individual array elements.

slicing: getting and setting smaller subarrays within a larger array.


Operaciones de suma importancia para operar con arreglos en Numpy.

a. **Atributos**

```python
x1 = np.random.randint(10, size=6)  # 1d
x2 = np.random.randint(10, size=(3,4))   # 2d
x3 = np.random.randint(10, size=(3,4,2))   #3d
# ndim --> numero de dimensiones
# shape --> forma
#size --> medida total
# dtype --> typo de los componentes
# itemsize --> dimension en bytes.
print(x1)
print('x3 ndim:', x3.ndim)
print('x3 shape:', x3.shape)
print('x3 size:', x3.size)
print('x3 dtype:', x3.dtype)
print('x3 itemsize:', x3.itemsize, 'bytes')
print('x3 nbytes:', x3.nbytes, 'bytes')  #size x itemsize
```

b. **Array indexing (acceso a elementos individuales en un arreglo por indice)**

c. **Array slicing(acceso a subarreglos)(:)**

th-3897625800.jpeg

```python
# X[star:stop:step]  : - caracteristica de slicing
x = np.arange(10)
print(x)
print(x[:8]) # primeros 7 elementos
print(x[5::3])  # elementos despues del 5to (4)
print(x[5:6])  # elementos entre el 5to y el 6mo
#x[::2] # elementos de a 2 comenzando en 0
#x[2::2] # de a 2 elementos comenzando en el indece 1
#x[::-1] # elementos de 1 empezando por el reverso
#x[::-3]
print(x[::-2]==x[-1::-2])
#x[-2::-2]
#x[5::-2] #  reverse los elemtos dea dos desde la 5ta posicion
#x[0::2]==x[::2]
```

```python
print(x2)   # combinacion de indexing (,) y slicing (:)
#x2[:3,:3] # extraer las 3 primeras filas y las 3 primeras columnas.
x2[:3,::2] # todas las filas y la 1 y 3  columnas
#x2[::-1,::-1] # desde atras iniciando en el elemento de la ultima fila y ultima columna
```

```python
print(x2)
#print(x2[1:,3]) # primera columna de x2 la transforma en fila
#print(x2[1,:])
print(x2[:,2]) # desecha todo excepto la 1 columna y la convierte en fila.
```
