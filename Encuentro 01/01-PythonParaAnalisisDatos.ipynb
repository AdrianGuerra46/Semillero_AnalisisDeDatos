#3. Sentencias con condicionales

Condicionales if, elif, else (Sentencia).
Creacion de una funcion (sentencia def). Creacion de un nuevo objeto.

```python
def sign_of_a(a): #declaracion de la funcion. programacion procedural.
  if a < 0.0:
    sign = 'negative'
  elif a > 0.0:
      sign = 'positive'
  else:
      sign = 'zero'
  return sign
a = 1.5
print('a is' ' ' + sign_of_a(a) )  #invocacion de la funcion.
a = 0.0
print('a is' ' ' + sign_of_a(a) )
```

```python
def suma(*art):    # tupla como argumento de una funcion. Argumentos varibles
    print(type(art))
    # <class 'tuple'>
    total = 0
    for n in art:
        total += n
    return total
suma(1, 3, 5, 11, 56)
```

```python
def suma(**kwargs):   #parametros en forma de llave y valor -diccionario
    suma = 0;
    for key, value in kwargs.items():
        print(key, value)
        suma += value
    return suma

suma(a=5, b=20, c=23, h=4, e=6)
```

```python
def suma(**kwargs):  #con diccionario
    suma = 0
    for key, value in kwargs.items():
        print(key, value)
        suma += value
    return suma
di = {'a': 10, 'b': 20, 'c': 10, 'd':12}
suma(**di)
```

#4. Funciones

```python
def derivates(f, x, h=0.0001):
  df = (f(x+h)-f(x-h))/(2.0*h)
  ddf = (f(x+h)-2.0*f(x)+f(x-h))/h**2
  return df, ddf

from math import atan #arctan
df1, ddf1 = derivates(atan, 2.0)
print('first derivative=', df1)
print('second derivative=', ddf1)
```

```python
def squares(a):
  for i in range(len(a)):
    a[i] = a[i]**2
a = [1, 2, 3, 4, 5]
squares(a)
print(a)
```
